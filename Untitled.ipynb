```python
total_runs_by_country = df.groupby('Team')['Runs'].sum()
sorted_countries = total_runs_by_country.sort_values(ascending=False)
print(sorted_countries)
```

```python
# Group by 'Player' and calculate the total runs for each player
total_runs_by_player = df.groupby('Player')['Runs'].sum()

# Sort the players based on total runs in descending order
sorted_players = total_runs_by_player.sort_values(ascending=False)

# Select the top 5 players
top_5_players = sorted_players.head(5)
print(sorted_players)
```

```python
# Group by 'Team', 'Player', and calculate the total runs for each player in each team
total_runs_by_player_team = df.groupby(['Team', 'Player'])['Runs'].sum()

# Sort the players based on total runs in descending order
sorted_players = total_runs_by_player_team.sort_values(ascending=False)

# Select the top 5 players
top_5_players = sorted_players.head(5)

# Display the top 5 players with the highest total runs including team names
print(top_5_players)
```

```python
# Group by 'Team', 'Player', and calculate the total runs for each player in each team
total_runs_by_player_team = df.groupby(['Team', 'Player'])['Runs'].sum()

# Sort the players based on total runs in descending order
sorted_players = total_runs_by_player_team.sort_values(ascending=False)

# Select the top 5 players
top_5_players = sorted_players.head(5)

# Define different colors for each bar
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'lightblue']

# Plot a bar chart for the top 5 players with different colors and rotated labels
plt.figure(figsize=(10, 6))
bars = top_5_players.plot(kind='bar', color=colors)
plt.title('Top 5 Players with Highest Total Runs in Each Team')
plt.xlabel('Team, Player')
plt.ylabel('Total Runs')

# Rotate x-axis labels by 45 degrees
plt.xticks(rotation=45, ha='right')

# Annotate each bar with its value
for i, v in enumerate(top_5_players):
    plt.text(i, v + 5, str(v), ha='center', va='bottom', color='black')

plt.show()
```

```python
# Sort the DataFrame by 'Wickets' in descending order
sorted_players_by_wickets = df.sort_values(by='Wickets', ascending=False)

# Select the top 5 players
top_5_players = sorted_players_by_wickets.head(5)

# Display the top 5 players with their country and wicket count
print("Top 5 Players with the Highest Number of Wickets:")
print(top_5_players[['Player', 'Team', 'Wickets']])
```

```python
player_names = top_5_players['Player']
wickets_counts = top_5_players['Wickets']
countries = top_5_players['Team']  # Replace with the actual column name containing country information

# Plotting the bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(player_names + ' (' + countries + ')', wickets_counts, color='lightcoral')
plt.title('Top 5 Players with Highest Number of Wickets')
plt.xlabel('Player (Country)')
plt.ylabel('Number of Wickets')

# Annotate each bar with its value
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black')

# Show the plot
plt.show()
```

```python
# Sort the DataFrame by 'Runs' in ascending order to get the top 10 players
top_10_players = df.sort_values(by='Runs', ascending=True).head(10)

# Bar chart for the top 10 players
plt.figure(figsize=(12, 8))
bars = plt.barh(top_10_players['Player'], top_10_players['Overs'], color='skyblue', label='Overs')
bars = plt.barh(top_10_players['Player'], top_10_players['Runs'], color='orange', label='Runs')

# Annotate each bar with its value
for bar, overs, runs, team in zip(bars, top_10_players['Overs'], top_10_players['Runs'], top_10_players['Team']):
    plt.text(bar.get_x() + bar.get_width(), bar.get_y() + bar.get_height() / 2,
             f'{overs} Overs, {runs} Runs\nTeam: {team}',
             ha='left', va='center', color='black')

plt.title('Top 10 Players - Overs and Runs (Bottom to Top)')
plt.xlabel('Count')
plt.ylabel('Player')
plt.legend()

plt.show()
```

```python
# Sort the DataFrame by 'Runs' given by each bowler in descending order to get the top 10 bowlers
top_10_bowlers = df.sort_values(by='Runs', ascending=False).head(10)

# Create a subplot with a specified size
fig, ax = plt.subplots(figsize=(12, 5))

# Bar chart for the top 10 bowlers
bars = ax.barh(top_10_bowlers['Player'], top_10_bowlers['Runs'], color='red', label='Runs Given')

# Annotate each bar with its value
for bar, runs in zip(bars, top_10_bowlers['Runs']):
    ax.text(bar.get_x() + bar.get_width(), bar.get_y() + bar.get_height() / 2,
            f'{runs} Runs Given',
            ha='left', va='center', color='black')

ax.set_title('Top 10 Bowlers - Runs Given (High to Low)', fontsize=16)
ax.set_xlabel('Runs Given', fontsize=12)
ax.set_ylabel('Bowler', fontsize=12)
ax.legend()

plt.show()
```

```python


# Select the top 30 players based on 'Overs'
df_player_overs_top_30 = df.sort_values(by='Overs', ascending=False).head(30)

fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot bars for 'Overs' on the primary y-axis (ax1)
index = np.arange(len(df_player_overs_top_30))
bars = ax1.bar(index, df_player_overs_top_30['Overs'], label='Overs', color='skyblue')

# Plot line for 'Maidens' on the same y-axis (ax1)
line = ax1.plot(index, df_player_overs_top_30['Maidens'], color='orange', marker='o', label='Maidens')

# Set labels and titles
ax1.set_xlabel('Player')
ax1.set_ylabel('Overs', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')  # Make sure y-axis labels are in blue

plt.title('Maiden Overs and Overs for the Top 30 Players')

# Combine legends
bars_legend = [bar for bar in bars]
line_legend = [line[0]]
plt.legend(bars_legend + line_legend, ['Overs', 'Maidens'], loc='upper left')

# Set x-axis ticks and labels
plt.xticks(index, df_player_overs_top_30['Player'], rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
wickets_by_team = df.groupby('Team')['Wickets'].sum().reset_index()

# Plot a bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(wickets_by_team['Team'], wickets_by_team['Wickets'], color='lightcoral')
plt.title('Wickets Taken by Each Team')
plt.xlabel('Team')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45, ha='right')

# Annotate each bar with its value
for bar, value in zip(bars, wickets_by_team['Wickets']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, str(value),
             ha='center', va='bottom', color='black')

plt.show()
```

```python
maiden_overs_by_team = df.groupby('Team')['Maidens'].sum().reset_index()

# Plot a bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(maiden_overs_by_team['Team'], maiden_overs_by_team['Maidens'], color='skyblue')
plt.title('Maiden Overs by Each Team')
plt.xlabel('Team')
plt.ylabel('Total Maiden Overs')

# Rotate x-axis labels by 45 degrees
plt.xticks(rotation=45, ha='right')

# Annotate each bar with its value
for bar, value in zip(bars, maiden_overs_by_team['Maidens']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, str(value),
             ha='center', va='bottom', color='black')

plt.show()
```

```python
run_rate_by_team = df.groupby('Team')['Run_rate'].mean().reset_index()

# Plot a bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(run_rate_by_team['Team'], run_rate_by_team['Run_rate'], color='lightgreen')
plt.title('Average Run Rate by Each Team (Bowlers)')
plt.xlabel('Team')
plt.ylabel('Average Run Rate')

# Rotate x-axis labels by 45 degrees
plt.xticks(rotation=45, ha='right')
# Annotate each bar with its value
for bar, value in zip(bars, run_rate_by_team['Run_rate']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01, f'{value:.2f}',
             ha='center', va='bottom', color='black')

plt.show()
```
